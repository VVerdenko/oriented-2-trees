```python
N_1 = [0] * 100
M = [0] * 100
N = [0] * 100

M_S = [0] * 100     # ряды в которые возводим экспоненту для нахождения M и M + 2N соответсвенно
N_S = [0] * 100

e = [0] * 100   # просто M(x) + 2*N(x) для удобства

s = [0] * 100

t = [0] * 100

sp = []      # sp[n][k] хранит список разбиений числа n на k слогаемых, необходимо чтобы возводить экспоненту в степень с помощью ряда Тейлора
f = []      # массив факториалов
```

```python
def find_f(max_n):
  f.append(1)
  for i in range(1, max_n+1):
    f.append(f[i-1] * i)

f = []
find_f(20)      # необходимо посчитать достаточно f и sp прежде, чем считать ряды
```

```python
def find_sp(max_n):
  sp.append([[[]]])
  for i in range(1, max_n + 1):
    sp.append([[[]]])
    for j in range(1, i + 1):
      sp[i].append([[]])
    sp[i][1] = [[i]]

  # разбиваем n на k + 1 слогаемых, припсывая i в конец разбияения n - i на k слогаемых

  for n in range(2, max_n + 1):
    for i in range(1, n):
      for k in range(1, n - i + 1):
        x = sp[n - i][k][:]
        for j in range(len(x)):
          x[j] = x[j] + [i]
        if sp[n][k + 1] == [[]]:
          sp[n][k + 1] = x
        else:
          sp[n][k + 1] += x

sp = []
find_sp(20)
```

```python
def find_N_1(n):
  sum = 0
  n -= 1
  for i in range(1, n):
    sum += e[i] * e[n - i]
  sum += 2 * e[n]

  N_1[n + 1] = 3 * sum
```

```python
def find_M(n):
  sum = 0
  # Сперва посчитаем ряд в который возведем экспоненту
  for k in range(1, n + 1):
    if n % (2 * k) == 0:
      x = n // (2 * k)
      sum += N_1[x] / k
  M_S[n] = sum

  sum = 0
  for k in range(1, n + 1):       # номер слогаемого в ряде Тейлора
    y = 0
    for u in sp[n][k]:            # какие степени взять от каждой скобки, чтобы за к скобок получить н
      x = 1
      for i in u:
        x *= M_S[i]
      y += x
    sum += y / f[k]

  M[n] = sum
```

```python
def find_N(n):
  sum = 0
  # Сперва посчитаем ряд в который возведем экспоненту
  for k in range(1, n + 1):
    if n % k == 0:
      x = n // k
      sum += 2 * N_1[x] / k
  N_S[n] = sum

  sum = 0
  for k in range(1, n + 1):        # номер слогаемого в ряде Тейлора
    y = 0
    for u in sp[n][k]:             # какие степени взять от каждой скобки, чтобы за к скобок получить н
      x = 1
      for i in u:
        x *= N_S[i]
      y += x
    sum += y / f[k]

  N[n] = (sum - M[n]) / 2
```

```python
def find_s(n):
  n -= 1
  sum = 0
  if n >= 3:
    for i in sp[n][3]:      # sp[n][3] как раз описывает разбиение n на 3 не нулевых слогаемых
      x = 1
      for j in i:
        x *= e[j]
      sum += x
  for i in range(1, n):
    sum += 3 * e[i] * e[n - i]
  sum += 3 * e[n]

  s[n + 1] = sum
```

```python
def main(max_n):
  M[1] = 0
  N[1] = 3
  N_1[1] = 3

  s[1] = 1

  e[1] = 6

  M_S[1] = 0
  N_S[1] = 6

  for i in range(2, max_n + 1):
    find_N_1(i)
    find_M(i)
    find_N(i)

    N_1[i] = round(N_1[i])
    M[i] = round(M[i])
    N[i] = round(N[i])

    e[i] = M[i] + 2 * N[i]

  for i in range(2, max_n + 1):
    find_s(i)
    s[i] = round(s[i])

  for i in range(1, max_n + 1):
    t[i] = N[i] + M[i] - 2 * s[i]

main(20)
print("N_1:", N_1)
print("M:  ", M)
print("N:  ", N)
print("s:  ", s)
```
